### Importing dataset and creating features and target variables

```python
dataset = pd.read_csv("Position_Salaries.csv")
```

```python
X = dataset.iloc[:,1:2].values
y = dataset.iloc[:,2].values
```


##### Training set is not used here, beacause the dataset is to small and we want to get a good prediction

from sklearn.model_selection import train_test_split
X_treino,X_teste,y_treino,y_teste = train_test_split(X, y, test_size=0.2, random_state=0)

### Fitting the Linear Regressor 

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

### Fitting the Polynomial Regressor

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
```

#### Create the polynomial matrix

```python
X_poly = poly_reg.fit_transform(X)
```

```python
lin_reg2 = LinearRegression()
```

```python
lin_reg2.fit(X_poly,y)
```

### Visualiazing Linear Regression Results

```python
plt.figure(figsize = (10,5))

plt.scatter(X, y , color ="red")
plt.plot(X, lin_reg.predict(X), color = "blue")
plt.title("Truth or Bluff (Linear Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

### Visualizing Polynomial Regression Results

#### Creating a high resolution view, with np.arange

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
```

```python
plt.figure(figsize = (10,7))

plt.scatter(X, y , color ="red")
plt.plot(X_grid, lin_reg2.predict(poly_reg.fit_transform(X_grid)), color = "blue")
plt.title("Truth or Bluff (Polynomial Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```
